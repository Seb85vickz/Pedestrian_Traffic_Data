```python
#Load CSV
print("Downloading CSV...", CSV_URL)
r = requests.get(CSV_URL)
r.raise_for_status()
df = pd.read_csv(io.BytesIO(r.content))
print("CSV loaded:", df.shape)
```

```python
#Parse list columns
list_cols = ['other_oid','other_class','other_x','other_y']
for c in list_cols:
    if c in df.columns and df[c].dtype == object:
        df[c] = df[c].apply(lambda x: ast.literal_eval(x) if pd.notna(x) and isinstance(x, str) else [])
df = df.sort_values(['oid','timestamp']).reset_index(drop=True)
```

```python
#1.Make a clean copy and sort
df_clean = df.copy()
df_clean = df_clean.sort_values(['oid', 'timestamp']).reset_index(drop=True)

#Orientation columns containing missing values
orient_cols = [
    'body_roll','body_pitch','body_yaw',
    'head_roll','head_pitch','head_yaw'
]

#2.Forward-fill and back-fill within each oid group
df_clean[orient_cols] = df_clean.groupby('oid')[orient_cols].transform(lambda g: g.ffill().bfill())

#3.Linear interpolation within each oid group
df_clean[orient_cols] = df_clean.groupby('oid')[orient_cols].transform(lambda g: g.interpolate(method='linear'))

#4.Global median fill (in case any column still has missing values)
for col in orient_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

#Check if any missing values remain
print(df_clean[orient_cols].isna().sum())
```

```python
#Interpolate orientation & compute gaze divergence
orient_cols = ['body_yaw','head_yaw','body_roll','body_pitch']
for col in orient_cols:
    if col in df.columns:
        df[col] = df.groupby('oid')[col].transform(lambda g: g.interpolate(method='linear', limit_direction='both'))
        df[col] = df[col].fillna(df[col].median())

def angdiff(a,b):
    d = np.abs(a-b) % 360
    return np.minimum(d, 360-d)

if 'head_yaw' in df.columns and 'body_yaw' in df.columns:
    df['gaze_divergence'] = angdiff(df['head_yaw'], df['body_yaw'])
    df['intention_state'] = df['gaze_divergence'].apply(lambda x: 'Scanning' if x>25 else 'Walking')
else:
    df['gaze_divergence'] = np.nan
    df['intention_state'] = 'unknown'
```

```python
from math import hypot

df_feat = df_clean.copy()

df_feat['num_neighbors'] = df_feat['other_oid'].apply(lambda x: len(x) if isinstance(x, (list,tuple)) else 0)

def safe_get_list(lst, idx, default=np.nan):
    try:
        return lst[idx]
    except Exception:
        return default

df_feat['first_neighbor_oid'] = df_feat['other_oid'].apply(lambda x: safe_get_list(x,0))
df_feat['n1_x'] = df_feat['other_x'].apply(lambda x: safe_get_list(x,0))
df_feat['n1_y'] = df_feat['other_y'].apply(lambda x: safe_get_list(x,0))

def centroid(xs):
    try:
        xs = list(xs)
        return float(np.mean(xs)) if len(xs)>0 else np.nan
    except Exception:
        return np.nan

df_feat['centroid_other_x'] = df_feat['other_x'].apply(centroid)
df_feat['centroid_other_y'] = df_feat['other_y'].apply(centroid)

# dx, dy and speed per oid by timestamp differences
df_feat[['dx','dy']] = df_feat.groupby('oid')[['x','y']].diff()
df_feat['dt'] = df_feat.groupby('oid')['timestamp'].diff()
df_feat['speed'] = np.sqrt(df_feat['dx']**2 + df_feat['dy']**2) / df_feat['dt']

df_feat['dist_to_centroid'] = np.sqrt((df_feat['x']-df_feat['centroid_other_x'])**2 + (df_feat['y']-df_feat['centroid_other_y'])**2)

df_feat[['num_neighbors','n1_x','n1_y','centroid_other_x','centroid_other_y','dx','dy','dt','speed','dist_to_centroid']].head()
```

```python
#Map calibration and world->pixel
resp_info = requests.get(MAP_INFO_URL)
map_info = {}
if resp_info.status_code == 200:
    for line in resp_info.text.splitlines():
        if ':' in line:
            k,v = line.split(':',1)
            k_stripped = k.strip()
            v_stripped = v.strip()
            if k_stripped == 'cellSize' and v_stripped.endswith('m'):
                try: map_info[k_stripped] = float(v_stripped[:-1])
                except: map_info[k_stripped] = v_stripped
            else:
                try: map_info[k_stripped] = float(v_stripped)
                except: map_info[k_stripped] = v_stripped
else:
    map_info = {'x_offset':495750.0,'y_offset':5405600.0,'cell_size':0.1}

def world_to_pixel(xw,yw,info):
    x_offset_key = 'X_offset' if 'X_offset' in info else 'x_offset'
    y_offset_key = 'y_offset'
    cell_size_key = 'cellSize' if 'cellSize' in info else 'cell_size'

    x_offset = float(info[x_offset_key])
    y_offset = float(info[y_offset_key])

    cell_size_val = info[cell_size_key]
    if isinstance(cell_size_val, str) and cell_size_val.endswith('m'):
        cell_size = float(cell_size_val[:-1])
    else:
        cell_size = float(cell_size_val)

    px = (xw - x_offset) / cell_size
    py = (yw - y_offset) / cell_size
    return px, py

if 'x' not in df.columns or 'y' not in df.columns:
    raise RuntimeError("tracks.csv must include 'x' and 'y' columns.")
df['pixel_x'], df['pixel_y'] = world_to_pixel(df['x'].values, df['y'].values, map_info)
```

```python
#Fetch map image
img = None
img_uri = None
img_w = img_h = None
resp_img = requests.get(MAP_IMG_URL)
if resp_img.status_code == 200 and 'image' in resp_img.headers.get('Content-Type',''):
    img = Image.open(BytesIO(resp_img.content)).convert('RGBA')
    img_w,img_h = img.size
    df['pixel_y_img'] = img_h - df['pixel_y']
    buf = BytesIO()
    img.save(buf, format='PNG')
    img_uri = 'data:image/png;base64,' + base64.b64encode(buf.getvalue()).decode('ascii')
    print('Loaded semantic map:', img_w, img_h)
else:
    df['pixel_y_img'] = df['pixel_y']
    print('Semantic map not available; proceeding without background image.')
```

```python
#Speeds & neighbors & duration
df['dx'] = df.groupby('oid')['x'].diff().fillna(0)
df['dy'] = df.groupby('oid')['y'].diff().fillna(0)
df['dt'] = df.groupby('oid')['timestamp'].diff().fillna(1)
df['speed_m_s'] = np.sqrt(df['dx']*2 + df['dy']*2) / df['dt']
df['neighbor_count'] = df['other_oid'].apply(lambda o: len(o) if isinstance(o, list) else 0)
dur = df.groupby('oid')['timestamp'].agg(['min','max']).reset_index()
dur['duration_s'] = dur['max'] - dur['min']
df = df.merge(dur[['oid','duration_s']], on='oid', how='left')
```

```python
#VIS 11: interaction graph snapshot at median timestamp
ts = int(df['timestamp'].median())
snap = df[df['timestamp']==ts]
G = nx.DiGraph()
for _, row in snap.iterrows():
    a = row['oid']; others = row.get('other_oid', [])
    if isinstance(others, list):
        for b in others:
            G.add_edge(str(a), str(b))
pos = nx.spring_layout(G, seed=1)
edge_x=[]; edge_y=[]
for e in G.edges():
    x0,y0 = pos[e[0]]; x1,y1 = pos[e[1]]
    edge_x += [x0,x1,None]; edge_y += [y0,y1,None]
node_x = [pos[n][0] for n in G.nodes()]; node_y = [pos[n][1] for n in G.nodes()]
fig = go.Figure()
fig.add_trace(go.Scatter(x=edge_x, y=edge_y, mode='lines', line=dict(width=1), hoverinfo='none'))
fig.add_trace(go.Scatter(x=node_x, y=node_y, mode='markers+text', text=list(G.nodes()), textposition='top center', marker=dict(size=10)))
fig.update_layout(title=f'Interaction graph at timestamp={ts}', width=800, height=600); fig.show()

# VIS 12: track duration distribution
dur_df = df.groupby('oid')['duration_s'].first().reset_index()
fig = px.histogram(dur_df, x='duration_s', nbins=40, title='Track duration distribution (s)')
fig.update_layout(width=700, height=450); fig.show()
```

1. Kinematic Features: Displacement (dx,dy), Time Delta(dt), Speed $\sqrt{dx^2 + dy^2} / dt$.

3. Contextual/Social Features: Neighbor Count (other_oid), Neighbor Centroids (centroid_other_x, centroid_other_y), Distance to Centroid (Eludian Distance)

4. Spatial Aggregation: Grid Cooridinates (gx,gy)
